```python
import pandas as pd

# Creating a manual dataset
data = {
    'customer_id': [101, 102, 103, 104, 105],
    'age': [25, 30, 35, 28, 40],
    'gender': ['Male', 'Female', 'Female', 'Male', 'Male'],
    'city': ['Delhi', 'Mumbai', 'Chennai', 'Kolkata', 'Pune'],
    'purchase_amount': [2000, 3000, 2500, 4000, 3500]
}

df = pd.DataFrame(data)
print(df)

# Alternative way to create the DataFrame using a list of lists
import pandas as pd

data = [
    [101, 25, 'Male', 'Delhi', 2000],
    [102, 30, 'Female', 'Mumbai', 3000],
    [103, 35, 'Female', 'Chennai', 2500],
    [104, 28, 'Male', 'Kolkata', 4000],
    [105, 40, 'Male', 'Pune', 3500]
]
columns = ['customer_id', 'age', 'gender', 'city', 'purchase_amount']

df = pd.DataFrame(data, columns=columns)
print(df)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Create a DataFrame
data = pd.DataFrame({
    'Age': [20, 25, 30, 35, 40],
    'Salary': [20000, 25000, 30000, 35000, 40000]
})

# Min-Max Scaling
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data)

# Print normalized data
print("Min-Max Normalized Data:\n", data_scaled)
```

```python
import pandas as pd

# Sample DataFrame
data = {
    'Gender': ['Male', 'Female', 'Female', 'Male'],
    'Department': ['IT', 'HR', 'IT', 'HR'],
    'Salary': [50000, 40000, 60000, 45000]
}
df = pd.DataFrame(data)

# One-Hot Encoding
encoded = pd.get_dummies(df, columns=['Gender'])
print("One-Hot Encoded DataFrame:")
print(encoded)

# GroupBy & Aggregation
result = df.groupby('Department')['Salary'].mean()
print("\nAverage Salary by Department:")
print(result)

# Min-Max Normalization (example on Salary)
df['Salary_Normalized'] = (df['Salary'] - df['Salary'].min()) / (df['Salary'].max() - df['Salary'].min())
print("\nNormalized Salary:")
print(df['Salary_Normalized'])
```
