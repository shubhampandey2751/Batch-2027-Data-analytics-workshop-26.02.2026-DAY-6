```python
import pandas as pd

data = {
    'customer_id': [101, 102, 103, 104, 105],
        'age': [25, 30, 35, 28, 40],
            'gender': ['Male', 'Female', 'Female', 'Male', 'Male'],
                'city': ['Delhi', 'Mumbai', 'Chennai', 'Kolkata', 'Pune'],
                    'purchase_amount': [2000, 3000, 2500, 4000, 3500]
                    }

df = pd.DataFrame(data)
print(df)


```

```python
import pandas as pd

data = [
    [101, 25, 'Male', 'Delhi', 2000],
            [102, 30, 'Female', 'Mumbai', 3000],
                        [103, 35, 'Female', 'Chennai', 2500],
                                        [104, 28, 'Male', 'Kolkata', 4000],
                                                            [105, 40, 'Male', 'Pune', 3500]
                                                                                ]

columns = ['customer_id', 'age', 'gender', 'city', 'purchase_amount']

df = pd.DataFrame(data, columns=columns)
print(df)
```
